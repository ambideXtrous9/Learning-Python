```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python


# Create a meshgrid of x1 and x2 values
x1 = np.linspace(-1, 1, 100)
x2 = np.linspace(-1, 1, 100)
X1, X2 = np.meshgrid(x1, x2)

# Calculate the function values based on the different regions
Z = np.abs(X1) + np.abs(X2)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z, cmap='viridis', alpha=0.8)

# Set axis labels
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('|x1| + |x2|')

# Show the plot
plt.show()
```

```python

# Generate synthetic data
np.random.seed(42)
X1 = 2 * np.random.rand(100, 1)
y = 4 + 3 * X1 + np.random.randn(100, 1)

# Create a meshgrid of w0 and w1 values
w0_vals = np.linspace(1, 7, 100)
w1_vals = np.linspace(2, 5, 100)
W0, W1 = np.meshgrid(w0_vals, w1_vals)

# Calculate the MSE values based on the linear regression model
m = len(y)
MSE = np.zeros_like(W0)

for i in range(len(w0_vals)):
    for j in range(len(w1_vals)):
        predictions = W0[i, j] + W1[i, j] * X1
        errors = predictions - y
        MSE[i, j] = np.sum(errors**2) / (2 * m)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W0, W1, MSE, cmap='viridis', alpha=0.8)

# Set axis labels
ax.set_xlabel('w0')
ax.set_ylabel('w1')
ax.set_zlabel('MSE')

# Show the plot
plt.title('MSE for y = w0 + w1*x1')
plt.show()
```
