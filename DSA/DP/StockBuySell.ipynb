## **Kadane's Algorithm : Maximum Subarray**

```python
import math
class Solution:
    def maxSubArray(self, nums) -> int:

        lm = 0
        gm = -math.inf

        for i in range(len(nums)):

            lm += nums[i]

            gm = max(lm,gm)

            if(lm < 0): lm = 0

        return gm
```

## **121. Best Time to Buy and Sell Stock**

#### **Kadane's Algorithm**

```python
import math

class Solution:
    def maxProfit(self, prices) -> int:

        stockBought = math.inf
        profit = 0

        for i in range(len(prices)):
            stockBought = min(stockBought,prices[i])
            profit = max(profit,prices[i]-stockBought)
        
        return profit     
```

#### **DP : Memoization**

```python
class Solution:
    
    def SBS(self,i,k,buy,n,P):

        if(i==n or k==0): return 0

        if(self.dp[i][k][buy]!=-1): return self.dp[i][k][buy]
        
        # Not buying or Selling on that day, skipping

        profit = self.SBS(i+1,k,buy,n,P)

        if(buy and k):
            
            # buying so paying money -> -ve

            profit = max(profit, -P[i] + self.SBS(i+1,k,not buy,n,P))

        else:
            
            # selling getting money -> +ve
            # k-1 -> 1 transaction is complete

            profit = max(profit, P[i] + self.SBS(i+1,k-1,not buy,n,P))

        self.dp[i][k][buy] = profit

        return self.dp[i][k][buy]




    def maxProfit(self, prices) -> int:

        n = len(prices)
        k = 1
        buy = True

        self.dp = [[[-1 for _ in range(3)] for _ in range(k+1)] for _ in range(n+1)]

        return self.SBS(0,k,buy,n,prices)
        
```

### **DP : Tabulation**

#### **Rules:**

**1. make indices reverse of Memoization:**

    in Memo : i -> n to 0
    in Tabu : i -> n to 0

    Memo : k -> 2 to 0
    Tabu : k -> 0 to 2

    Memo : buy -> 1 to 0
    Tabu : buy -> 0 to 1
           
2. then same conditions write with idxs

3. to tackle return -> check if else properly

4. Get the results at finishing idx : dp[0][k = 1][buy = 1]

```python
class Solution:

    def maxProfit(self, prices) -> int:

        n = len(prices)
        k = 1
        buy = True
        TransacFee = 0
        cooldown = 0

        dp = [[[0 for _ in range(3)] for _ in range(k+1)] for _ in range(n+1)]


        for i in range(n,-1,-1):

            for j in range(k+1):

                for t in range(buy+1):

                    if(i==n or j==0): 
                        
                        dp[i][j][t] = 0

                    else:

                        profit = dp[i+1][j][t]

                        if(j and t):

                            profit = max(profit, -prices[i] + dp[i+1][j][not t])
                        
                        else:

                            profit = max(profit, prices[i] - TransacFee + dp[i+1+cooldown][j-1][not t])
                        
                        dp[i][j][t] = profit
                

        return dp[0][k][buy]
        
```

## **122. Best Time to Buy and Sell Stock II**

**Infinite Transactions**

```python
class Solution:
    def maxProfit(self, prices) -> int:
        
        n = len(prices)
        # k = 1
        buy = True
        TransacFee = 0
        cooldown = 0

        dp = [[0 for _ in range(3)] for _ in range(n+1)]

        for i in range(n,-1,-1):

            for j in range(buy+1):

                if(i==n): 
                    
                    dp[i][j] = 0

                else:

                    profit = dp[i+1][j]

                    if(j):

                        profit = max(profit, -prices[i] + dp[i+1][not j])
                    
                    else:

                        profit = max(profit, prices[i] - TransacFee + dp[i+1+cooldown][not j])
                    
                    dp[i][j] = profit
                

        return dp[0][buy]
        
```

## **309. Best Time to Buy and Sell Stock with Cooldown**

```python
class Solution:
    def maxProfit(self, prices) -> int:
        
        n = len(prices)
        # k = 1
        buy = True
        TransacFee = 0
        cooldown = 1

        dp = [[0 for _ in range(3)] for _ in range(n+1+cooldown)]

        for i in range(n,-1,-1):

            for j in range(buy+1):

                if(i==n): 
                    
                    dp[i][j] = 0

                else:

                    profit = dp[i+1][j]

                    if(j):

                        profit = max(profit, -prices[i] + dp[i+1][not j])
                    
                    else:

                        profit = max(profit, prices[i] - TransacFee + dp[i+1+cooldown][not j])
                    
                    dp[i][j] = profit
                

        return dp[0][buy]
        
```
