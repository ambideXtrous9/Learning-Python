## **64. Minimum Path Sum**



```python
import math

class Solution:

    def DFS(self,i,j,n,m,grid):

        if(i>=n or j>=m):
            return math.inf
        
        if(i==n-1 and j==m-1): return grid[i][j]


        if(self.dp[i][j]!=-1): return self.dp[i][j]
        

        self.dp[i][j] = grid[i][j] + min(self.DFS(i+1,j,n,m,grid),self.DFS(i,j+1,n,m,grid))

        return self.dp[i][j]

    def minPathSum(self, grid) -> int:

        n = len(grid)
        m = len(grid[0])

        self.dp = [[-1]*(m+1) for _ in range(n+1)]

        return self.DFS(0,0,n,m,grid)

```

```python
import math
class Solution:

    def minPathSum(self, grid) -> int:

        n = len(grid)
        m = len(grid[0])

        dp = [[math.inf]*(m+1) for _ in range(n+1)]

        for i in range(n,-1,-1):

            for j in range(m,-1,-1):

                if(i>=n or j>=m): 
                    dp[i][j] = math.inf

                elif(i==n-1 and j==m-1): 
                    dp[i][j] = grid[i][j]

                else:

                    dp[i][j] = grid[i][j] + min(dp[i+1][j],dp[i][j+1])

        return dp[0][0]
```
