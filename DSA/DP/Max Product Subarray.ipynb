## **152. Maximum Product Subarray**

1. Same intuition as Kadane's Algorithm (R -> L and L -> R)

2. We have to calculate this for both side : Prefix and Suffix



```python
import math

class Solution:
    def maxProduct(self, nums) -> int:

        maxpdt = -math.inf
        curpdt = 1

        n = len(nums)

        # Prefix
        for x in nums:

            curpdt *= x

            maxpdt = max(maxpdt,curpdt)

            if(curpdt == 0): 
                curpdt = 1

        curpdt = 1

        # Suffix
        for i in range(n-1,-1,-1):

            curpdt *= nums[i]

            maxpdt = max(maxpdt,curpdt)

            if(curpdt == 0): 
                curpdt = 1

        return maxpdt 
```
