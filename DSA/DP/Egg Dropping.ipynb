## **1884. Egg Drop With 2 Eggs and N Floors**





### Base Cases





```python
import math
class Solution:
    def EGD(self,f,e):

        if(f == 0 or f == 1) : return f

        if(e == 1) : return f

        if(self.dp[f][e]!=-1) : return self.dp[f][e]

        res = math.inf

        for k in range(1,f+1):

            ways = 1 + max(self.EGD(k-1,e-1), self.EGD(f-k,e))

            res = min(ways,res)

        self.dp[f][e] = res

        return self.dp[f][e]

    def twoEggDrop(self, n: int) -> int:

        k = 2

        dp = [[-1]*(k+1) for _ in range(n+1)]

        for i in range(n+1):

            for j in range(1,k+1):

                if(i == 0 or i==1): dp[i][j] = i

                elif(j == 1) : dp[i][j] = i

                else:

                    res = math.inf

                    for x in range(1,i+1):

                        ways = 1 + max(dp[x-1][j-1],dp[i-x][j])

                        res = min(res,ways)
                    
                    dp[i][j] = res

        return dp[n][k]

        return self.EGD(n,k)
```
