## **1143. Longest Common Subsequence**



```python
class Solution:
    
    def LCS(self,s1,s2,n,m):

        if(n==0 or m==0): return 0

        if(self.dp[n][m]!=-1): return self.dp[n][m]

        if(s1[n-1] == s2[m-1]):

            self.dp[n][m] = 1 + self.LCS(s1,s2,n-1,m-1)

        else:

            self.dp[n][m] = max(self.LCS(s1,s2,n,m-1),self.LCS(s1,s2,n-1,m))

        return self.dp[n][m]



    def longestCommonSubsequence(self, s1: str, s2: str) -> int:

        n = len(s1)
        
        m = len(s2)

        dp = [[0]*(m+1) for _ in range(n+1)]


        for i in range(n+1):

            for j in range(m+1):

                if(i==0 or j==0): dp[i][j] = 0

                elif(s1[i-1] == s2[j-1]):
                    dp[i][j] = 1 + dp[i-1][j-1]
                
                else:

                    dp[i][j] = max(dp[i-1][j],dp[i][j-1])

        return dp[n][m]
        
```

## **72. Edit Distance**



```python
class Solution:
    

    def ED(self,s1,s2,n,m):

        if(n==0): return m

        if(m==0): return n

        if(self.dp[n][m]!=-1): return self.dp[n][m]

        if(s1[n-1] == s2[m-1]): 

            self.dp[n][m] = self.ED(s1,s2,n-1,m-1)
        
        else:

            self.dp[n][m] = 1 + min((self.ED(s1,s2,n-1,m),self.ED(s1,s2,n,m-1),self.ED(s1,s2,n-1,m-1)))

        return self.dp[n][m]


    def minDistance(self, s1: str, s2: str) -> int:

        n = len(s1)
        m = len(s2)

        dp = [[0]*(m+1) for _ in range(n+1)]


        for i in range(n+1):

            for j in range(m+1):

                if(i==0) : dp[i][j] = j

                elif(j==0) : dp[i][j] = i

                elif(s1[i-1] == s2[j-1]):

                    dp[i][j] = dp[i-1][j-1]

                else:

                    dp[i][j] = 1 + min((dp[i-1][j],dp[i][j-1],dp[i-1][j-1]))

        return dp[n][m]
```
