## **312. Burst Balloons**






```python
import math
class Solution:

    def BB(self,l,r,nums):

        if(l>r) : return 0

        if(self.dp[l][r] != -1) : return self.dp[l][r]

        res = -math.inf

        for k in range(l,r+1):

            score = nums[l-1] * nums[k] * nums[r+1] + self.BB(l,k-1,nums) + self.BB(k+1,r,nums)
            res = max(res,score)

        self.dp[l][r] = res
        
        return self.dp[l][r]

    def maxCoins(self, nums: List[int]) -> int:

        n = len(nums)
        nums.insert(0,1)
        nums.append(1)

        dp = [[0]*(n+2) for _ in range(n+2)]

        for l in range(n,0,-1):

            for r in range(l,n+1):

                if(l>r) : dp[l][r] = 0

                else:
                    res = -math.inf

                    for k in range(l,r+1):

                        score = nums[l-1] * nums[k] * nums[r+1] + dp[l][k-1] + dp[k+1][r]

                        res = max(res,score)
                    
                    dp[l][r] = res

        return dp[1][n]

        return self.BB(l,r,nums)
```
