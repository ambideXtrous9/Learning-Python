## **MCM**



```python
class Solution:
    
    def MCM(self,i,j,arr):
        
        if(i>=j): return 0
        
        if(self.dp[i][j]!=-1): return self.dp[i][j]
        
        res = float('inf')
        
        for k in range(i,j):
            
            calc = arr[i-1]*arr[k]*arr[j] + self.MCM(i,k,arr) + self.MCM(k+1,j,arr)
            
            res = min(res,calc)
            
            
        self.dp[i][j] = res
        
        return self.dp[i][j]
    
    def matrixMultiplication(self, N, arr):
        
        dp = [[0]*(N+1) for _ in range(N+1)]
        
        
        for i in range(N-1,0,-1):
            
            for j in range(0,N): # or we can replace with i+1 to N then no need to check (i>=j)
                
                if(i >= j): 
                    
                    dp[i][j] = 0
                
                else:
                    
                    res = float('inf')
                    
                    for k in range(i,j):
                        
                        calc = arr[i-1] * arr[k] * arr[j] + dp[i][k] + dp[k+1][j]
                        
                        res = min(res,calc)
                        
                    dp[i][j] = res
                    
        
        return dp[1][N-1]
                    
```

## **1547. Minimum Cost to Cut a Stick**




```python
import math

class Solution:

    def Cut(self,sstick,estick,lidx,ridx,cuts):

        if(lidx > ridx): return 0

        if(self.dp[lidx][ridx]!=-1): return self.dp[lidx][ridx]

        res = math.inf

        for i in range(lidx,ridx+1):

            temp = (estick-sstick) + (self.Cut(sstick,cuts[i],lidx,i-1,cuts) +
                                      self.Cut(cuts[i],estick,i+1,ridx,cuts))
            res = min(res,temp)

        self.dp[lidx][ridx] = res

        return self.dp[lidx][ridx] 


    def minCost(self, n: int, cuts) -> int:

        estick = n
        sstick = 0

        cuts.sort()

        lidx = 0
        ridx = len(cuts)-1

        self.dp = [[-1]*(ridx+1) for _ in range (ridx+1)]

        return self.Cut(sstick,estick,lidx,ridx,cuts)
    
```
