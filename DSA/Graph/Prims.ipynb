```python
import heapq as hq
import math


class Solution:
    
    def spanningTree(self, V, adj):
        
        s = 0 # start node
        
        key = [math.inf for _ in range(V)] # store optimal weights
        
        mSet = [False for _ in range(V)] # node already in MST
        
        parent = [-1 for _ in range(V)] # parent of every node that are in MST, required to print the MST
        
        parent[s] = -1 
        
        key[s] = 0
        
        Q = []
        
        hq.heappush(Q,(key[0],s)) # Min Heap to pick min edge every time
        
        while Q:
            
            (cost,u) = hq.heappop(Q)
            
            mSet[u] = True   # node is included in the MST
            
            for item in adj[u]:
                
                v = item[0] # v : adjacent node of u
                
                weight = item[1] # distance of v from u
                
                if(mSet[v] == False and key[v]>weight):  # saves from forming cycle
                    
                    key[v] = weight            # edge W is updated with min val
                    
                    parent[v] = u
                    
                    hq.heappush(Q,(key[v],v))
                    
        
        res = 0
        for x in key: res += x
        
        return res
```
