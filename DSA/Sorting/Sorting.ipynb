## **Bubble Sort**



### **1. at every step largest element will end up at the end of the array,**
### **2. and i will reduce from the end ==> (n-1) to 0**
### **3. j will iterate from start to < i for every i in the inner loop.**
### **4. and if arr[j] > arr[j+1] ==> then swap** 

![image2.png](https://miro.medium.com/v2/resize:fit:1400/0*dnWHML2AavZiJEVE.gif)

````python
import time
````

````python
arr = [5,3,6,28,3,9,0,5,3]
````

````python
def BubbleSorting(arr):
    
    n = len(arr)
    
    for i in range(n-1,-1,-1):
        for j in range(0,i):
            if(arr[j] > arr[j+1]):
                arr[j],arr[j+1] = arr[j+1],arr[j]
    return arr
````

````python
start = time.time()
arr = BubbleSorting(arr=arr)
end = time.time()
print("Sorted List = ",arr)
print("Time Elapsed = ",(end-start))
````

# Selection Sort

- **i**: current position to be updated

- **j**: starts from **i + 1** to search for the number to be placed at **i** in each iteration.

### Process
1. **j** starts at **(i + 1)**. Therefore, **i** will reach till **n - 1** and **j** will iterate till **n**.

2. For every position starting from 0, find the smallest number for that position in every iteration and swap.

3. **i** indicates the boundary between the sorted (left) and unsorted (right) parts.

4. **i** starts from 0 and goes till **n - 1**.

5. **j** starts from **i + 1** and goes till **n**.

6. Swap the elements if needed.

### Pseudocode
```cpp
min = i;
for (j = i + 1; j < n; j++) {
    if (arr[min] > arr[j]) 
        min = j;
}
swap(arr[i], arr[min]);


````python
def SelectionSort(arr):
    
    n = len(arr)
    
    for i in range(0,n-1):
        
        minidx = i
        
        for j in range(i+1,n):
            
            if(arr[j] < arr[minidx]) : 
                minidx = j
                
        arr[i],arr[minidx] = arr[minidx],arr[i]
        
    return arr
````

````python
start = time.time()
arr = SelectionSort(arr=arr)
end = time.time()
print("Sorted List = ",arr)
print("Time Elapsed = ",(end-start))
````

## **Insertion Sort**




### Process
1. Select an element in each iteration from the unsorted portion of the array (using an outer loop).
2. Place this element in its corresponding position within the sorted part by shifting the remaining elements as necessary (using an inner loop and swapping).
3. The inner `while` loop shifts elements by swapping, creating the correct order.

-------------------------------------------------------------

1. At each iteration, the left part of **i** will be sorted, while the right part remains unsorted.
2. For each iteration, **j** will traverse from **i** to 0 and swap the elements if **arr[j-1] > arr[j]**.

### Pseudocode
```cpp
for (i = 0; i < n; i++) {
    j = i;
    while (j > 0 && arr[j-1] > arr[j]) {
        swap(arr[j-1], arr[j]);
        j--;
    }
}
```

![imagee.png](https://blog.shahadmahmud.com/wp-content/uploads/2020/04/is3.gif)

````python
def InsertionSort(arr):
    
    n = len(arr)
    
    for i in range(0,n):
        
        j = i
        
        while(j > 0 and arr[j] < arr[j-1]):
            
            arr[j],arr[j-1] = arr[j-1],arr[j]
            
            j -= 1
            
    return arr
````

````python
start = time.time()
arr = InsertionSort(arr=arr)
end = time.time()
print("Sorted List = ",arr)
print("Time Elapsed = ",(end-start))
````

## **Merge Sort**

![img.png](https://codepumpkin.com/wp-content/uploads/2017/10/MergeSort_Avg_case.gif) ![img2.png](https://i.imgur.com/HU2tfzo.gif) 



````python
def merge(arr,low,mid,high):
    
    temp = []
    
    left = low
    right = mid+1
    
    while(left <= mid and right <= high):
        
        if(arr[left] > arr[right]):
            
            temp.append(arr[right])
            right += 1
        else:
            temp.append(arr[left])
            left += 1
            
            
    while(left <= mid) : 
        temp.append(arr[left])
        left += 1
        
    while(right <= high):
        temp.append(arr[right])
        right += 1
        
    for i in range(low,high+1):
        
        arr[i] = temp[i-low]
````

````python
def mergeSort(arr,low,high):
    
    if(low >= high) : return
    
    mid = (high-low)//2 + low
    
    mergeSort(arr,low,mid)
    mergeSort(arr,mid+1,high)
    merge(arr,low,mid,high)
````

````python
def mergeSortUtil(arr):
    
    n = len(arr)
    
    low = 0
    high = n-1
    
    mergeSort(arr,low,high)
    
    return arr
````

````python
start = time.time()
arr = mergeSortUtil(arr=arr)
end = time.time()
print("Sorted List = ",arr)
print("Time Elapsed = ",(end-start))
````

## **QuickSort**

 

![x.png](https://blog.shahadmahmud.com/wp-content/uploads/2020/04/qs1-1.gif)

````python
arr = [5,3,6,28,3,9,0,5,3]
````

````python
def partition(arr,low,high):
    
    pivot = arr[high]
    
    pi = low
    
    for i in range(low,high+1):
        
        if(arr[i] < pivot):
            
            arr[pi],arr[i] = arr[i],arr[pi]
            
            pi += 1
            
    arr[pi],arr[high] = arr[high],arr[pi]
    return pi
    
````

````python
def quickSort(arr,low,high):
    
    if low > high : return
    
    pivot = partition(arr,low,high)
    
    quickSort(arr,low,pivot-1)
    quickSort(arr,pivot+1,high)
````

````python
def quickSortUtil(arr):
    
    n = len(arr)
    low = 0
    high = n-1
    
    quickSort(arr,low,high)
````

````python
start = time.time()
quickSortUtil(arr=arr)
end = time.time()
print("Sorted List = ",arr)
print("Time Elapsed = ",(end-start))
````
