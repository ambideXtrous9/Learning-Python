## **Largest BST in BT**



1. The left and right subtrees of every tree node are BST.

2. A nodes value should be more than the largest value in the left subtree 
and less than the smallest value in the right subtree.

**at every Node we have 2 such array from L and R :**

**{Largest from SubTree, Smallest from SubTree, isBST, size}**

```python
import math

class Node:
    def __init__(self,val,left=None,right=None):
        self.data = val
        self.left = left
        self.right = right
        
class Solution:
        
    def __init__(self):    
        self.res = 0
        
    def LGBST(self,root):
        
        if root == None: 
            return [-math.inf,math.inf,True,0]
        
        L = self.LGBST(root.left)
        R = self.LGBST(root.right)
        
        isBST = False
        size = 0
        
        if(L[2] and R[2] and L[0] < root.data and R[1] > root.data):
            isBST = True
        
        if(isBST):
            size = 1 + L[3] + R[3]
            
        maxFromCur = max(root.data,R[0])
        minFromCur = min(root.data,L[1])
        
        self.res = max(self.res,size)
        
        return [maxFromCur,minFromCur,isBST,size]
        
    def largestBst(self, root):
        
        self.LGBST(root)
        
        return self.res
```

## **Floor in BST**

```python
class Solution:
    
    def BST(self,root,x):
        
        if(root == None) : return
    
        if(root.data <= x):
            self.ans = root.data
            self.BST(root.right,x)
            
        else:
            self.BST(root.left,x)
            
    
    def floor(self, root, x):
        
        self.ans = -1
        
        self.BST(root,x)
        
        return self.ans
        
```

## **108. Convert Sorted Array to Binary Search Tree**

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
class Solution:

    def BT(self,nums,l,r):

        if(l>r): return None

        mid = (r-l)//2 + l

        n = nums[mid]

        root = TreeNode(val=n)

        root.left = self.BT(nums,l,mid-1)

        root.right = self.BT(nums,mid+1,r)

        return root

    def sortedArrayToBST(self, nums: List[int]) -> Optional[TreeNode]:

        l = 0
        r = len(nums)-1

        return self.BT(nums,l,r)
        
```

## **700. Search in a Binary Search Tree**

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
        
class Solution:

    def S(self,root,t):

        if(not root): return None

        if(root.val == t) : return root

        if(root.val < t) : return self.S(root.right,t)

        else:

            return self.S(root.left,t)

    def searchBST(self, root: Optional[TreeNode], val: int) -> Optional[TreeNode]:

        return self.S(root,val)
        
```

## **100. Same Tree**



```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
class Solution:

    def ST(self,u,v):

        if(not u and not v): return True

        if(not u and v) : return False

        if(u and not v) : return False

        return (u.val == v.val) and self.ST(u.left,v.left) and self.ST(u.right,v.right)


    def isSameTree(self, p: Optional[TreeNode], q: Optional[TreeNode]) -> bool:

        return self.ST(p,q)
```

## **Height Balanced Binary Tree**

### **Brute Force**

```python
class Solution:
    # Function to check if a binary tree is balanced
    def isBalanced(self, root):
        # If the tree is empty, it's balanced
        if not root:
            return True
        
        # Calculate the height of left and right subtrees
        leftHeight = self.getHeight(root.left)
        rightHeight = self.getHeight(root.right)

        # Check if the absolute difference in heights
        # of left and right subtrees is <= 1
        if abs(leftHeight - rightHeight) <= 1 \
            and self.isBalanced(root.left) \
            and self.isBalanced(root.right):
            return True

        # If any condition fails, the tree is unbalanced
        return False
```

### **Optimal**


```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:

    def HBT(self,root):

        if(not root): return 0

        lh = self.HBT(root.left)
        rh = self.HBT(root.right)

        if(lh == -1) : return -1

        if(rh == -1) : return -1

        if(abs(rh-lh) > 1) : return -1

        return 1 + max(lh,rh)

    def isBalanced(self, root: Optional[TreeNode]) -> bool:

        return True if self.HBT(root)!=-1 else False
```

## **103. Binary Tree Zigzag Level Order Traversal**

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

        
class Solution:
    def zigzagLevelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:

        if(not root): return []

        LOR = []

        Q = []

        Q.append(root)

        flag = 1

        while Q:

            temp = []

            for i in range(len(Q)):

                node = Q.pop(0)

                temp.append(node.val)

                if(node.left) : Q.append(node.left)

                if(node.right) : Q.append(node.right)
            
            if(flag): LOR.append(temp)
            if(not flag): LOR.append(temp[::-1])

            flag = (not flag)


        return LOR
```

## **Check for Children Sum Property in a Binary Tree**

```python
class Node:
    def init(self,val):
        self.data = val
        self.left = None
        self.right = None

class Solution:
    #Function to check whether all nodes of a tree have the value 
    #equal to the sum of their child nodes.
    def SS(self,root):
        
        if(not root): return True # True for None
        
        if(not root.left and not root.right) : return True # True for single Node
        
        lc = root.left.data if (root.left) else 0
        rc = root.right.data if (root.right) else 0
        
        csum = lc + rc
        
        return (csum == root.data) and self.SS(root.left) and self.SS(root.right)
        
    def isSumProperty(self, root):
        # code here
        
        return 1 if (self.SS(root)==True) else 0
```

## **1008. Construct Binary Search Tree from Preorder Traversal**

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class Solution:

    def BBT(self,l,r,pre):

        if(l>r): return None

        node = pre[self.idx]

        inidx = self.M[node]

        self.idx += 1

        root = TreeNode(val=node)

        root.left = self.BBT(l,inidx-1,pre)

        root.right = self.BBT(inidx+1,r,pre)

        return root

    def bstFromPreorder(self, preorder: List[int]) -> Optional[TreeNode]:

        inorder = sorted(preorder)

        n = len(inorder)

        self.M = {}

        self.idx = 0

        for i in range(n):

            self.M[inorder[i]] = i

        return self.BBT(0,n-1,preorder)
        
```
