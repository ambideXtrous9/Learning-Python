## **Largest BST in BT**



1. The left and right subtrees of every tree node are BST.

2. A nodes value should be more than the largest value in the left subtree 
and less than the smallest value in the right subtree.

**at every Node we have 2 such array from L and R :**

**{Largest from SubTree, Smallest from SubTree, isBST, size}**

```python
import math

class Node:
    def __init__(self,val,left=None,right=None):
        self.data = val
        self.left = left
        self.right = right
        
class Solution:
        
    def __init__(self):    
        self.res = 0
        
    def LGBST(self,root):
        
        if root == None: 
            return [-math.inf,math.inf,True,0]
        
        L = self.LGBST(root.left)
        R = self.LGBST(root.right)
        
        isBST = False
        size = 0
        
        if(L[2] and R[2] and L[0] < root.data and R[1] > root.data):
            isBST = True
        
        if(isBST):
            size = 1 + L[3] + R[3]
            
        maxFromCur = max(root.data,R[0])
        minFromCur = min(root.data,L[1])
        
        self.res = max(self.res,size)
        
        return [maxFromCur,minFromCur,isBST,size]
        
    def largestBst(self, root):
        
        self.LGBST(root)
        
        return self.res
```

## **Floor in BST**

```python
class Solution:
    
    def BST(self,root,x):
        
        if(root == None) : return
    
        if(root.data <= x):
            self.ans = root.data
            self.BST(root.right,x)
            
        else:
            self.BST(root.left,x)
            
    
    def floor(self, root, x):
        
        self.ans = -1
        
        self.BST(root,x)
        
        return self.ans
        
```
