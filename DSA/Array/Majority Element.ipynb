## **The Boyer-Moore Majority Vote Algorithm**

1. Firstly we will declare a count variable to keep track of the count and an element variable to store the potential majority element.

2. We will start iterating through the array and if the count is 0 set the value of the element equal to the current element and increment count.

3. If the current element is equal to the element, increment count; otherwise, decrement count.











4. After the iteration element is the potential element to verify it we will run a loop to check whether the potential element is the actual majority element or not. 

```
count = 0
for x in nums: 
    if(x==elem): count += 1

if count > (n//2) : return elem

```


````python
import math

class Solution:
    def majorityElement(self, nums) -> int:

        count = 0
        elem = math.inf

        n = len(nums)

        for x in nums:

            if count == 0:

                count += 1
                elem = x

            elif(x == elem): 
                
                count += 1

            # No Match
            else: 
                count -= 1
            
            #print(elem,end=" ")
        
        count = 0

        for x in nums: 
            if(x==elem): count += 1

        if count > (n//2) : return elem

        return -1
    
nums = [2,1,2,1,1,1,3]

obj = Solution()

obj.majorityElement(nums)
        
````

## **229. Majority Element II**

````python
import math

class Solution:
    
    def majorityElement(self, nums):

        cnt1 = 0
        cnt2 = 0

        elem1 = math.inf
        elem2 = math.inf

        for x in nums:

            if(cnt1 == 0 and x != elem2):

                cnt1 += 1
                elem1 = x
            
            elif(cnt2 == 0 and x != elem1):

                cnt2 += 1
                elem2 = x
            
            elif(x == elem1): cnt1 += 1

            elif(x == elem2): cnt2 += 1

            # No match
            else:

                cnt1 -= 1
                cnt2 -= 1

        
        cnt1 = 0
        cnt2 = 0

        for x in nums:

            if x == elem1: cnt1 += 1
            if x == elem2: cnt2 += 1

        res = []

        mini = (len(nums)//3) 

        if(cnt1 > mini): res.append(elem1)
        if(cnt2 > mini): res.append(elem2)

        return res
````
