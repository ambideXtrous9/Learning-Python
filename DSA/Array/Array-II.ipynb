## **287. Find the Duplicate Number**

**Given an array of integers nums containing n + 1 integers where each integer is in the range [1, n] inclusive.**

#### **1. Using Set**

```python
class Solution:
    def findDuplicate(self, nums) -> int:

        H = set()

        for x in nums:

            if x in H: return x

            else: H.add(x)
        
```

#### **Pigeon Hole Principle : Binary Search**



```python
class Solution:
    def findDuplicate(self, nums) -> int:

        n = len(nums)

        low = 0
        high = n-1

        while(low<high):

            mid = (high-low)//2 + low

            cnt = 0

            for x in nums: 
                if(x <= mid): cnt += 1

            if(cnt <= mid) : low = mid +1

            else: high = mid
        
        return low
```

#### **2. Floyd's Cycle Dtetction : Slow Fast Pointer**

1. Convert the array to LinkedList
2. Floyd's Cycle Dtetction using Slow Fast Pointer



```python
class Solution:
    def findDuplicate(self, nums) -> int:

        slow = 0
        fast = 0

        while(True):

            slow = nums[slow]
            fast = nums[nums[fast]]

            if(slow == fast): break

        slow = 0

        while(True):

            slow = nums[slow]
            fast = nums[fast]

            if(slow == fast): break

        return slow
```
