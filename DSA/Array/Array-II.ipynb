### **Transpose of a Matrix**

```python
def Matrix(M):
    
    n = len(M)
    m = len(M[0])
    
    for i in range(n):
        
        for j in range(i+1,n):
            
            M[i][j],M[j][i] = M[j][i],M[i][j]
            
```

```python
def Matrix(M):
    
    n = len(M)
    m = len(M[0])
    
    for i in range(n):
        
        for j in range(i):
            
            M[i][j],M[j][i] = M[j][i],M[i][j]
    
    for X in M: X.reverse()
    
    print(M)
            
```
