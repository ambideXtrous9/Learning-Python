## **62. Unique Paths**



#### **Recursion + Memoization**

```python
class Solution:
    
    def DFS(self,n,m):

        if(m==1 and n==1): return 1
        
        if(n<=0 or m<=0): return 0

        if(self.dp[n][m]!=-1): return self.dp[n][m]

        self.dp[n][m] =  self.DFS(n-1,m) + self.DFS(n,m-1)

        return self.dp[n][m]
    
    
    def uniquePaths(self, m: int, n: int) -> int:

        self.dp = [[-1]*(m+1) for _ in range(n+1)]

        return self.DFS(n,m)


        
```

#### **Tabulation:**

```python
class Solution:
    
    def DFS(self,n,m):

        dp = [[0]*(m+1) for _ in range(n+1)]

        for i in range(n+1):

            for j in range(m+1):

                if(i<=0 or j<=0): 
                    
                    dp[i][j] = 0
                
                if(i==1 and j==1):

                    dp[i][j] = 1
                
                else:

                    dp[i][j] = dp[i-1][j] + dp[i][j-1]

        return dp[n][m]

    
    def uniquePaths(self, m: int, n: int) -> int:

        return self.DFS(n,m)
```
