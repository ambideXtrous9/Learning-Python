## **128. Longest Consecutive Sequence**



```python
from collections import defaultdict

class Solution:
    def longestConsecutive(self, nums) -> int:
        
        D = defaultdict(int)

        for x in nums:

            D[x] = 1

        for x in D.keys():    
        
            if (x-1) in D:
        
                D[x] = 0

        mxlen = 0

        for x in D.keys():

            if D[x] == 0: continue

            temp = 0

            while(x in D):

                temp += 1
                x += 1

            mxlen = max(mxlen,temp) 

        return mxlen
```

```python
class Solution(object):
    def longestConsecutive(self, nums):
        
        mySet = set(nums)

        if(len(nums) == 0): return 0
        if(len(nums) == 1): return 1

        longest = 1

        res = [i for i in nums if(i-1) not in mySet]

        for i in res:
            
            temp = 1

            while((i + 1) in mySet):
                temp += 1
                i += 1

            longest = max(longest,temp)            

        return longest
```
