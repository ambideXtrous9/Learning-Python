## **Dutch National Flag : Sort 0,1,2**

### **75. Sort Colors**



### **Algorithm to get the Dutch National Flag pattern:**

1. **When curr points to a 0**, we swap it with the element at zero_ptr. After swapping we move curr and zero_ptr one step forward.

    **swap(A[zeroptr],A[curptr]); zeroptr = ++; curptr = ++**

2. **When curr points to a 2**, we swap it with the element at two_ptr. We then decrement the value of two_ptr. Note that we should not move curr at this step because we may have just swapped a 2 with another 2 and we should re-process it in the next step.

    **swap(A[twoptr],A[curptr]); twoptr = --; curptr = unchanged**


3. **If curr points to a 1**, we just move curr to the next element.

    **curptr = ++**

4. Repeat till **(cur <= twoptr)** 





```python
class Solution:
    def sortColors(self, A) -> None:
        
        n = len(A)
        cur = 0
        ptr0 = 0

        ptr2 = (n-1)

        while(cur <= ptr2):

            if(A[cur] == 0):
                A[cur],A[ptr0] = A[ptr0],A[cur]
                cur += 1
                ptr0 += 1

            elif(A[cur] == 2):
                A[cur],A[ptr2] = A[ptr2],A[cur]
                ptr2 -= 1

            elif(A[cur] == 1):
                cur += 1
            
        return A
```
