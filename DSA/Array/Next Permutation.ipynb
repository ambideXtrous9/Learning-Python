## **31. Next Permutation**

![ChessUrl](https://nicklee1006.github.io/Leetcode-31-Next-Permutation/example.gif "NP")







```python
class Solution:
    def nextPermutation(self, nums) -> None:
        
        n = len(nums)

        idx = -1

        if(n==0): return []

        if(n==1): return nums
        
        for i in range(n-2,-1,-1):

            if(nums[i] < nums[i+1]): 
                idx = i
                break

        if(idx == -1): return nums.reverse()

        else:

            for j in range(n-1,idx,-1):

                if(nums[j] > nums[idx]):

                    nums[idx],nums[j] = nums[j],nums[idx]
                    
                    break

            nums[idx+1:] = reversed(nums[idx+1:])

            return nums

        
```
